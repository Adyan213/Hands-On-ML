```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data="https://github.com/ageron/data/raw/main/"
```

```python
life_sat=pd.read_csv(data+"lifesat/lifesat.csv")
```

```python
X=life_sat[['GDP per capita (USD)']].values
y=life_sat[['Life satisfaction']].values
```

```python
life_sat.plot(kind="scatter", x="GDP per capita (USD)", y="Life satisfaction", grid=True)
plt.axis([23_500, 62_500, 4, 9])
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
X_new=[[37_655.2]]
```

```python
print(model.predict(X_new))
```

```python
life_sat.plot(kind="scatter", x="GDP per capita (USD)", y="Life satisfaction", grid=True)
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Life satisfaction")

# Plot the linear regression line
X_plot = np.linspace(23_500, 62_500, 100).reshape(-1, 1)
y_plot = model.predict(X_plot)
plt.plot(X_plot, y_plot, color='red', label='Linear Regression')

plt.axis([23_500, 62_500, 4, 9])
plt.legend()
plt.show()
```
